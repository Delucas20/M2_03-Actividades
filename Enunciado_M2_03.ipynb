**1) Utiliza el método input() para realizar las siguientes tareas:**

* Pide un número por teclado, muestra el resultado y su tipo (para conocer en que tipo de dato captura la información input())
* Pide un número por teclado y asegurate de que capturemos la información en formato int
* Pide un número por teclado y asegurate de que capturemos la información en formato float

__Comprobar con type que los datos son del tipo correcto__

```python

print("=== Ejercicio 1.1: Número y su tipo ===")
numero_texto = input("Introduce un número: ")
print(f"Valor: {numero_texto}")
print(f"Tipo: {type(numero_texto)}")


print("\n=== Ejercicio 1.2: Número como int ===")
numero_int = int(input("Introduce un número entero: "))
print(f"Valor: {numero_int}")
print(f"Tipo: {type(numero_int)}")


print("\n=== Ejercicio 1.3: Número como float ===")
numero_float = float(input("Introduce un número decimal: "))
print(f"Valor: {numero_float}")
print(f"Tipo: {type(numero_float)}")
```

**2) Vamos a formatear números:**

* Pide un numero entero por teclado
* Muéstralo con el formato de 5 dígitos rellenos con 0s por delante
* Pide un número flotante por teclado
* Muéstralo con 5 dígitos para la parte entera y 3 dígitos para la parte decimal

```python

numero_entero = int(input("Introduce un número entero: "))
print(f"Número formateado (5 dígitos con 0s): {numero_entero:05d}")

numero_flotante = float(input("Introduce un número decimal: "))
print(f"Número formateado (5 dígitos enteros, 3 decimales): {numero_flotante:08.3f}")
```

**3) Vamos a ponernos creativos. ¿Cuántas formas tienes de mostrar esta información?**<bR>
<code>La altura es de 1,80 metros y el peso es de 80,135 KG</code>
    
Siendo la altura y el peso dos variables que se recojan con 2 inputs
    
__Pista: format() se puede usar de varias formas__

```python

altura = float(input("Introduce la altura (metros): "))
peso = float(input("Introduce el peso (kg): "))

print("\n=== Forma 1: f-string ===")
print(f"La altura es de {altura} metros y el peso es de {peso} KG")


print("\n=== Forma 2: format() ===")
print("La altura es de {} metros y el peso es de {} KG".format(altura, peso))


print("\n=== Forma 3: format() con nombres ===")
print("La altura es de {h} metros y el peso es de {p} KG".format(h=altura, p=peso))


print("\n=== Forma 4: Concatenación ===")
print("La altura es de " + str(altura) + " metros y el peso es de " + str(peso) + " KG")


print("\n=== Forma 5: Formato % ===")
print("La altura es de %s metros y el peso es de %s KG" % (altura, peso))


print("\n=== Forma 6: f-string con decimales ===")
print(f"La altura es de {altura:.2f} metros y el peso es de {peso:.3f} KG")
```
